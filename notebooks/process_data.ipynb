```python
from datasets import load_dataset

ds = load_dataset("nvidia/OpenMathInstruct-2", split = "train")
```

```python
ds
```

```python

def tokenize_function(examples):
    from transformers import AutoTokenizer

    tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-Math-1.5B-Instruct")

    messages = [[
        {"role": "user", "content": problem},
        {"role": "assistant", "content": solution}
    ] for problem, solution in zip(examples["problem"], examples["generated_solution"])]
    formatted_texts = tokenizer.apply_chat_template(
        messages, add_generation_prompt=False, tokenize=False
    )
    tokenized = tokenizer(
        formatted_texts, max_length=4096, truncation=True,
        padding="max_length", return_tensors="pt"
    )
    tokenized["labels"] = tokenized["input_ids"].clone()

    return tokenized
```

```python

def create_prompt_tokenize_function(examples):
    from transformers import AutoTokenizer

    tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-Math-1.5B-Instruct")

    messages = [[
        {"role": "user", "content": problem},
        {"role": "assistant", "content": solution}
    ] for problem, solution in zip(examples["problem"], examples["generated_solution"])]
    formatted_texts = tokenizer.apply_chat_template(
        messages, add_generation_prompt=False, tokenize=False
    )
    return {
        "text": formatted_texts,
    }
```

```python
non_tokenized_ds = ds.map(
    create_prompt_tokenize_function,
    batched=True,
    remove_columns=ds.column_names,
    desc="Creating prompt text",
    num_proc=16
)
```

```python
# Tokenize the dataset
tokenized_ds = ds.map(
    tokenize_function,
    batched=True,
    remove_columns=ds.column_names,
    desc="Tokenizing dataset", num_proc=8
)
```

```python
tokenized_ds.push_to_hub(
    "vinhpx/nvidia-math-dataset",
    split="tokenized"
)
```

```python
non_tokenized_ds.push_to_hub(
    "vinhpx/nvidia-math-dataset",
    split="non_tokenized"
)
```
